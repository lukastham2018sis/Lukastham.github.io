```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
patient_df = pd.read_csv("PatientInfo.csv")
```

```python
new_df = patient_df.dropna(subset=['deceased_date'])
```

```python
my_dict = {}
for index, row in new_df.iterrows():
    if row.age not in my_dict:
        my_dict[row.age] = 1
    else:
        my_dict[row.age] += 1
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = []

students = []
for key in my_dict:
    langs.append(key)
langs.sort()
for key in my_dict:
    students.append(my_dict[key])
students.sort()

ax.bar(langs,students)
plt.show()
```

```python
gender_dict = {}
for index, row in new_df.iterrows():
    if row.sex not in gender_dict:
        gender_dict[row.sex] = 1
    else:
        gender_dict[row.sex] += 1
print(gender_dict)
```

```python
#Probability
#P(M): 
total_gender = sum(gender_dict.values())
male_death = gender_dict['male'] / total_gender
female_death = gender_dict['female'] / total_gender
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
gender = []
num = []
for key in gender_dict:
    gender.append(key)
gender.sort()
for key in gender_dict:
    num.append(gender_dict[key])
num.sort()

ax.bar(gender,num)
plt.show()
```

```python
healthy_patient = patient_df[~patient_df.state.str.contains('deceased')]
healthy_patient = healthy_patient[~healthy_patient.state.str.contains('isolated')]
healthy_patient = healthy_patient.dropna(subset=['sex'])
```

```python
from sklearn import preprocessing
gender = []
infec = []
classes = []
le = preprocessing.LabelEncoder()
for sex in patient_df.sex:
    gender.append(sex)
for infection_case in patient_df.infection_case:
    infec.append(infection_case)
for state in patient_df.state:
    classes.append(state)
sex_encoded = le.fit_transform(gender)
infection_case_encoded = le.fit_transform(infec)
classes_encoded = le.fit_transform(classes)
```

```python
print(sex_encoded)
#1 = male
#0 = female
```

```python
print(infection_case_encoded)

#23 oversea inflow
#18 contact with patient
#15 Seongdong-gu APT
#19 etc
#4  Eunpyeong St. Mary's Hospital
#16 Shincheonji Church
#3 Dongan Church
#6 Guro-gu Call Center
#12 Onchun Church
#2 Cheongdo Daenam Hospital
```

```python
print(classes_encoded)
#2 released
#1 isolated
#0 death
```

```python
# features = zip(sex_encoded[0:2],infection_case_encoded[0:2])
features = [i for i in zip(sex_encoded, infection_case_encoded)]
```

<h1>Naive Bayes Analysis</h1>
<p> It is observed that males whose infectious case came from Cheongdo Daenam Hospital have the highest probability of death, while females have the higher probability of being released if infectious case came from the same source </p>

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(features,classes_encoded)

predicted = model.predict([[1,2]])
print(predicted)
```
